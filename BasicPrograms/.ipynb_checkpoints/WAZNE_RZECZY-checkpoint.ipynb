```python
# KLASY


#PRZYKADOWE KLASY  ( TUTAJ S ZE SOB POWIZANE )

# skadowe klast wykorzystujemu po self   (  self.zmienna )  ale do funkcji jak wida podajemy normalnie
# S to metody ktre maj z gry okrelone kiedy zostan wywoaane
# DOC - string  -  czyli taka !!!    dokumentacja    !!! do klasy ktra bdzie si wywietla w maym okienku przy tekscie 
# kiedy najedziemy na nazw klasy przy tworzeniu obiektu.Dziaa to te w przypadku metod klasy a nie tyklo dla samej klasy , ponadto z rozszerzeniem autoDOCstream 
# mona odrazu tworzy te dokumentacj dla poszczeglnych zmiennych
# Metody d-under to te ktre maj dwa podkrelniki np.__init__
# __str__  -  to nalee interpretowa jako " co si stanie/ma si sta   jeli obiekt zostanie potraktorany jako string "
# __getitem__ - to metoda ktra Dziaa kiedy chcemy wycign jakby element obiektu po indeksie  i moemy to ustawi 
#  eby zwracao wtedy to o co nam chodzi 
# __setitem__ - to metoda ktra troche jakby czya si z getitem i po wybraniu indexu zmienia wybrany przez naz atrybut
#  czyli przypisuje co do wybranej pozyscji

# stare rakiety

class Rocket:
    """Klasa Rakieta
    """
    def __init__(self,hight:float=0.0,power:float=1.0) -> None:
        """Konstruktor

        Args:
            hight (float, optional): _description_. Defaults to 0.0.
            power (float, optional): _description_. Defaults to 1.0.
        """
        self.hight = hight
        self.power = power
    
    def __str__(self) -> str:
        return "Rakieta o mocy " + str(self.power) + " znajduje si na wysokoci " + str(self.hight)


    def move_up(self) -> None:
        """Metoda do poruszania si do gry w zalerznoci od mocy rakiety
        """
        self.hight += 1 * self.power   #liniowy wzrost
        #self.hight = random.randint(1,10) * self.power  # randomowy wzrost wysokoci na podstawie mocy




class BlockOfRockets:
    def __init__(self,AmmoutOfRockets:int=3) -> None:
        """Class which creates a block of rockets

        Args:
            AmmoutOfRockets (int, optional): Ammount of rocketst that we want to have in our block . Defaults to 3.
        """
        self.AmmountOfRockets = AmmoutOfRockets
        self.rockets = [Rocket(i,i) for i in range(self.AmmountOfRockets)]

    def MoveBlock(self):
        """ Moves rocket up accordingly to power of the specyfic rocket
        """
        for rocket in self.rockets:
            rocket.hight += 1 * rocket.power

    def __getitem__(self,key):
        return self.rockets[key]


    def __str__(self) -> str:
        """ Print info about Block of rockets
        """
        Print = ''
        for Rocket in self.rockets:
            Print += Rocket.__str__() + '\n'
        return Print


BlockR = BlockOfRockets(4)

BlockR.MoveBlock
print(BlockR)

```
